## 1. 📖 Historia y evolución de YOLO <a name="historia"></a>

**YOLO (You Only Look Once)** es uno de los modelos de detección de objetos y segmentación de imágenes más populares del mundo. Fue desarrollado originalmente por **Joseph Redmon y Ali Farhadi** en la Universidad de Washington. Lanzado en **2015**, YOLO ganó popularidad por su alta velocidad y precisión.

### Línea de tiempo de versiones

| Versión | Año | Innovación principal |
|---------|-----|----------------------|
| **YOLOv1** | 2015 | Arquitectura original — detección en tiempo real |
| **YOLOv2** | 2016 | Batch normalization, anchor boxes, dimension clusters |
| **YOLOv3** | 2018 | Backbone más eficiente, múltiples anchors, SPP |
| **YOLOv4** | 2020 | Mosaic augmentation, anchor-free head, nueva función de pérdida |
| **YOLOv5** | 2020 | Optimización de hiperparámetros, experiment tracking, auto-export |
| **YOLOv6** | 2022 | Open-source por Meituan, usado en robots de entrega autónomos |
| **YOLOv7** | 2022 | Estimación de pose en COCO keypoints dataset |
| **YOLOv8** | 2023 | Ultralytics: flexibilidad total, soporte para todas las tareas de visión IA |
| **YOLOv9** | 2024 | PGI (Programmable Gradient Information) y GELAN |
| **YOLOv10** | 2024 | Universidad de Tsinghua — End-to-End head, elimina NMS |
| **YOLO11** | Sep 2024 | SOTA en detección, segmentación, pose, tracking y clasificación |
| **YOLO26** | 2025 | Optimizado para edge deployment, inferencia NMS-free |

> 📚 **Referencia oficial:** [https://docs.ultralytics.com/es](https://docs.ultralytics.com/es)

## 5. 📁 Estructura del dataset <a name="dataset"></a>

### Pasos para preparar un dataset personalizado

1. **Recopilar imágenes:** Reúna las imágenes del conjunto de datos (bases de datos públicas, Roboflow, captura propia)
2. **Anotar imágenes:** Agregue bounding boxes, segmentos o keypoints según la tarea
3. **Exportar anotaciones:** Convierta a formato YOLO `.txt` compatible con Ultralytics
4. **Organizar el dataset:** Use la siguiente estructura de carpetas:

```
dataset/
├── train/
│   ├── images/
│   └── labels/
├── valid/
│   ├── images/
│   └── labels/
└── test/
    ├── images/
    └── labels/
```

5. **Crear `data.yaml`:** En el directorio raíz del dataset, cree el archivo de configuración:

```yaml
path: /content/dataset          # ruta raíz del dataset
train: ./train/images
val: ./valid/images
test: ./test/images

nc: 2                           # número de clases
names: ['sin_casco', 'con_casco']  # nombres de las clases
```

6. **Optimizar imágenes (opcional):** Reducir resolución o comprimir para acelerar el entrenamiento
7. **Comprimir el dataset:** Crear un `.zip` para subir a Colab o compartir

### Variantes de modelos YOLOv8 disponibles

| Modelo | Tamaño | Velocidad | Precisión |
|--------|--------|-----------|----------|
| `yolov8n.pt` | Nano | ⚡⚡⚡⚡⚡ | ⭐⭐ |

### Parámetros principales de entrenamiento
- `data`: Ruta al archivo `.yaml` del dataset
- `epochs`: Número de épocas de entrenamiento
- `imgsz`: Tamaño de imagen de entrada (ej. 640)
- `batch`: Tamaño del lote (`-1` = auto según memoria GPU)
- `name`: Nombre del experimento
- `project`: Nombre del proyecto

> 📖 Más parámetros en: [Documentación Ultralytics](https://docs.ultralytics.com/modes/train/)

````python
# Mover todos los archivos exportados a /content/modelos_exportados/
import shutil, os

weights_dir = "/content/runs/ppe_detector/weights"
export_dir = "/content/modelos_exportados"

def get_dir_size(path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if os.path.exists(fp): # Check if file exists before getting size
                total_size += os.path.getsize(fp)
    return total_size

if os.path.exists(weights_dir):
    for f in os.listdir(weights_dir):
        src_path = os.path.join(weights_dir, f)
        dst_path = os.path.join(export_dir, f)

        if os.path.isfile(src_path):
            shutil.copy2(src_path, dst_path)
        elif os.path.isdir(src_path):
            if os.path.exists(dst_path):
                shutil.rmtree(dst_path) # Remove existing directory to allow clean copy
            shutil.copytree(src_path, dst_path)

    print("📦 Archivos copiados a", export_dir)
    for f in os.listdir(export_dir):
        item_path = os.path.join(export_dir, f)
        if os.path.isfile(item_path):
            size = os.path.getsize(item_path) / 1e6
            print(f"  {f:<40} {size:.1f} MB")
        elif os.path.isdir(item_path):
            size = get_dir_size(item_path) / 1e6
            print(f"  {f:<40} {size:.1f} MB (Directorio)")
else:
    print("⚠️ No se encontró la carpeta de pesos. Complete el entrenamiento primero.")
````
